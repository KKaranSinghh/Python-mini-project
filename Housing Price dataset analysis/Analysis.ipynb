```python
# New column of total numbers of rooms in house

df['Total rooms']= df['bedrooms'] + df['bathrooms']
df
```

```python
# Number of duplicate rows

df.duplicated().sum()
```

```python
# Check missing / null values

df.isnull().sum()
```

```python
# Sort on the basis of price of houses in ascending order

df.sort_values(by= 'price', ascending= True)
df
```

```python
# Distribution of Furnishingstatus between houses 

plt.figure(figsize= (8, 6))
df['furnishingstatus'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['skyblue', 'lightblue', 'pink'])  ###
plt.title('Distribution of furnishing status of houses')
plt.show()
```

```python
# Box plot for price distribution by number of bedrooms

plt.figure(figsize=(10, 6))
sns.boxplot(x='bedrooms', y='price', data= df)
plt.title('Price Distribution by Number of Bedrooms')
plt.xlabel('Number of Bedrooms')
plt.ylabel('Price')
plt.show()
```

```python
# Relationship between area and house price

plt.figure(figsize= (10, 6))
plt.scatter(df['area'], df['price'], color= 'lightblue', edgecolors= 'black') ###
plt.title('Relationship between area and house price')
plt.xlabel('Area')
plt.ylabel('House price')
plt.show()
```

```python
# Plot the bar chart
# Group the data by mainroad
grouped= df.groupby('mainroad')['price']

grouped.plot(kind= 'bar', color= ['lightblue', 'red'])
plt.title('Average house price based on main road access')
plt.xlabel('Main road access')
plt.ylabel('Avereg house price')
plt.xticks([0,1],['No', 'Yes'], rotation= 0)  ###
plt.show()
```

```python
# Group by hotwaterheating and airconditioning and calculate the average price

grouped_data= df.groupby(['hotwaterheating','airconditioning'])['price'].mean()

# Create a bar chart

plt.figure(figsize=(10,6))
grouped_data.unstack().plot(kind='bar', color= ['skyblue', 'pink'])
plt.title('Average house price by hot water heating and air conditioning')
plt.xlabel('Hot water hating / Air conditioning')
plt.ylabel('Average price')
plt.legend(title= 'Air conditioning', labels= ['No', 'Yes'])
plt.xticks(rotation= 0)
plt.show()
```
